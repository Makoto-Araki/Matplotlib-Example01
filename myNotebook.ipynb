```python
# ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib # 凡例等で日本語表記時に必要

# X軸とY軸の値
X = np.linspace(-20, 20, 100) # -20から20までの値を100個の値に等間隔で分割
Y1 = 20 * X + 130
Y2 = 2 * X * X

# X値とY値の範囲を指定してグラフ表示範囲を制限
plt.xlim(-15, 15)
plt.ylim(-50, 300)

# グラフタイトルを表示
plt.title('二次関数と一次関数のグラフ描画', fontsize=20)

# X軸とY軸のラベル設定
plt.xlabel('X値(-15:15)', fontsize=12)
plt.ylabel('Y値(-50:300)', fontsize=12)

# グリッド線を表示
plt.grid(True)

# X軸とY軸の目盛のパラメータ設定
plt.tick_params(labelsize=12)

# グラフ描画
plt.plot(X, Y1, color='g', label='一次関数')
plt.plot(X, Y2, color='r', label='二次関数')
plt.legend(loc="lower right", fontsize=12) # 凡例を表示
plt.show()
```
